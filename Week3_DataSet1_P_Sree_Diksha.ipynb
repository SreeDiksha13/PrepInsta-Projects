# **Dataset 1** Data Cleaning

```python
import numpy as np
import pandas as pd
```

```python
data_frame= pd.read_csv("/content/Data-cleaning-for-beginners-using-pandas.csv")
```

```python
data_frame.head()
```

```python
data_frame.shape
```

```python
data_frame.describe()  #Describing will help to find if data is missing in any column. As we can see, Age has missing values
```

```python
data_frame.info()  #We can find the data types of each column
```

```python
data_frame.isna().sum() #Here also we can see Age and Rating have missing values
```

```python
data_frame.columns
```

```python
data_frame.columns = data_frame.columns.str.lower().str.replace(" ", "_")
data_frame
```


**Cleaning Age Column**

```python
age_mean= data_frame["age"].mean()
```

```python
data_frame["age"]= data_frame["age"].fillna(age_mean).round(decimals=1)
data_frame
```

```python
data_frame["age"].isna().sum()
```

```python
data_frame["age"]= data_frame["age"].astype(int)
```

**Cleaning Salary Column**

```python
data_frame["salary"]= data_frame["salary"].apply(lambda x:x.replace("$", "").replace("k", "000"))
```

```python
data_frame["salary"]
```

```python
data_frame.head()
```

**Cleaning Rating Column**

```python
data_frame["rating"].isna().sum()
```

```python
#We can also observe some -1 values which is not valid, cleaning up that first:
```

```python
data_frame["rating"]= data_frame["rating"].replace(-1, 0)
```

```python
rating_mean= data_frame["rating"].mean()
data_frame["rating"]= data_frame["rating"].fillna(rating_mean)
```

```python
data_frame["rating"].head()
```

```python
data_frame["rating"].isna().sum()
```

```python
data_frame["rating"].round(decimals=2)
```

**Cleaning the Location Column**

```python
data_frame["location"].head()
```

```python
data_frame["location"]=data_frame["location"].str.replace("Autralia Aus", "Australia, Aus")
data_frame["location_city"]=data_frame["location"].str.split(",", expand=True).get(0)
data_frame["city_sign"]=data_frame["location"].str.split(",", expand=True).get(1)
data_frame.drop("location", inplace=True, axis=1)
```

```python
data_frame.head()
```

**Cleaning the established column**

```python
data_frame["established"]= data_frame["established"].replace(-1, np.nan)
data_frame["established"]  #We have replaced the -1 with nan and then nan with unknown as avg or median cannot for taken for this column
```

```python
data_frame["established"].fillna("Unknown", inplace=True)
data_frame["established"]
```

**Cleaning the Easy Apply Column**

```python
data_frame["easy_apply"]= data_frame["easy_apply"].replace("-1", "False").replace("TRUE", "True")
```

```python
data_frame["easy_apply"]
```

```python
data_frame["easy_apply"]= data_frame["easy_apply"].astype(bool) #changing the type of data in Easy apply to bool to classify as True or False.
```
